```python
import csv
import pandas as pd
import os
```

```python
df_train = pd.read_csv(r'Dataset\grambeddings_dataset_main\train.csv')
df_test = pd.read_csv(r'Dataset\grambeddings_dataset_main\test.csv')
```

```python
print(df_train.columns)
print(df_test.columns)
```

```python
df_train.drop('Unnamed: 0',axis=1, inplace=True)
df_test.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df_train.to_csv(r'Dataset\grambeddings_dataset_main\train.csv', index=False)
df_test.to_csv(r'Dataset\grambeddings_dataset_main\test.csv',index=False)
```

```python
rows = []
with open(r'Dataset\grambeddings_dataset_main\test.csv', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        # if more than 2 fields, merge all after the first one into URL
        if len(row) > 2:
            row = [row[0], ','.join(row[1:])]
        rows.append(row)

df4_train = pd.DataFrame(rows, columns=['label', 'url'])
df4_train.to_csv(r'Dataset\grambeddings_dataset_main\test.csv')
```
